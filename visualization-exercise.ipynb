```python
import matplotlib.pyplot as plt
```

```python
 cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta', 
          'Surakarta','Surabaya', 'Medan', 'Makassar')
 
populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
 
plt.bar(x=cities, height=populations)
plt.show()
```

```python
plt.bar(x=cities, height=populations)
plt.xticks(rotation=45)
plt.show()
```

membuat bar secara horizontal

```python
plt.barh(y=cities, width=populations)
plt.show()
```

kita boleh mengurutkan nilainya berdasarkan jumlah populasi terbanyak. kita membutuhkan bantuan library pandas dalam mengubah data menjadi DataFrame dan mengurutkannya menggunakan function sort_values().

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Cities': cities,
    'Population': populations,
})
```

```python
df
```

```python
df.sort_values(by='Population', inplace=True)
```

```python
df
```

```python
plt.barh(y=df["Cities"], width=df["Population"])
plt.show()
```

```python
plt.barh(y=df["Cities"], width=df["Population"])
plt.xlabel("Population (Millions)")
plt.ylabel("Name of the cities")
plt.title("Population of Cities in Indonesia")
plt.show()
```

membuat bar chart menggunakan library seaborn

```python
import seaborn as sns
```

```python
sns.barplot(y=df["Cities"], x=df["Population"], orient="h", color='blue')
plt.xlabel("Population (Millions)")
plt.ylabel("Name of the cities")
plt.title("Population of Cities in Indonesia")
plt.show()
```

```python
flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.1, 0, 0, 0)

plt.pie(
    x=votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)
plt.show()
```

```python
plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    wedgeprops = {'width': 0.4}
)
plt.show()
```

membuat grafik histogram untuk sumbu X data kuantitatif

```python
import numpy as np
```

```python
x = np.random.normal(15, 5, 250)
 
plt.hist(x=x, bins=15)
plt.show()
```

membuat grafik histogram dengan library seaborn

```python
sns.histplot(x=x, bins=15)
plt.show()
```

```python
sns.histplot(x=x, bins=15, kde=True)
plt.show()
```

untuk mengidentifikasi distribusi pada data kuantitatif ialah box plot. Ia merupakan bentuk visual untuk merepresentasikan nilai IQR beserta ambang batas bawah dan atas dari sebuah data. Box plot ini juga bisa digunakan untuk mengidentifikasi outlier atau pencilan yang ada pada data kuantitatif.

```python
x = np.random.normal(15, 5, 250)
sns.boxplot(x)
plt.show()
```
